```python
import pandas as pd
import os 
import re
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from tqdm import tqdm
import tensorflow as tf
import tensorflow_hub as hub
```

```python
base_path = '/home/roram/Data/datasets/qanda'
```

```python
data = pd.read_csv(os.path.join(base_path,'qanda_episodes.csv'))

data['hashtags'] = data['hashtags'].fillna('').apply(lambda s: s.split(';;;'))

data['mentions'] = data['mentions'].fillna('').apply(lambda s: s.split(';;;'))

data['urls'] = data['urls'].fillna('').apply(lambda s: s.split(';;;'))

data['text_ht_censored'] = data['text'].apply(lambda t: re.sub(r'http\S+', '<URL>', t)).apply(lambda t: re.sub("#[A-Za-z0-9_]+","<HASHTAG>", t)) 
```

```python
hashtag_labels = pd.read_csv(os.path.join(base_path,'archive','hashtags_labelled_100.csv'))
def assign_polarity(e):
    if(e in {'left', 'more left', 'left?', 'left_'}):
        return 1
    elif(e in ['right', 'more right', 'right?']):
        return -1
    else:
        return np.nan

hashtag_labels['polarity'] = hashtag_labels['label'].apply(assign_polarity)
hashtag_labels = hashtag_labels[~hashtag_labels['polarity'].isna()]
```

```python
left_hashtags = hashtag_labels[hashtag_labels['polarity'] == 1]['hashtags']
right_hashtags = hashtag_labels[hashtag_labels['polarity'] == -1]['hashtags']

def ground_truth_hashtag(ht):
    if ht in left_hashtags.values:
        return -1
    elif ht in right_hashtags.values:
        return 1
    else:
        return np.nan
```

```python
data_user_grouped_hashtags = data.groupby('uid').apply(lambda d: [e for l in d['hashtags'] for e in l])

user_ground_truth = data_user_grouped_hashtags.apply(lambda l: np.nanmean(list(map(ground_truth_hashtag, l))))

user_ground_truth_df = pd.DataFrame({'hashtags' : data_user_grouped_hashtags[data_user_grouped_hashtags.apply(len) > 0]})
user_ground_truth_df['continuous_gt_score'] = user_ground_truth
def handle_labels(s):
    if np.isnan(s):
        return -1
    elif s > 0:
        return 2
    elif s == 0:
        return 0
    elif s < 0:
        return 1
user_ground_truth_df['gt_label'] = user_ground_truth_df['continuous_gt_score'].apply(handle_labels)
```

```python
def dummy_fun(doc):
    return doc

tfidf = TfidfVectorizer(
            analyzer='word',
            tokenizer=dummy_fun,
            preprocessor=dummy_fun,
            token_pattern=None,
            min_df = 10)

toy = data_user_grouped_hashtags[data_user_grouped_hashtags.apply(len) > 0]

ht_embeddings = tfidf.fit_transform(toy.apply(lambda l: [str(s) for s in l])).todense()
```

```python
module_url = "https://tfhub.dev/google/universal-sentence-encoder/4"
model = hub.load(module_url)
print ("module %s loaded" % module_url)
def embed(input):
    return model(input)
```

```python
def batched_embed(l_text):
    all_embeddings = []
    chunk_size = 10
    for i in tqdm(range(0, len(l_text), chunk_size)):
        chunk = l_text[i: min(i+chunk_size, len(l_text))]
        emb = embed(chunk)
        all_embeddings.append(emb.numpy())
    return np.vstack(all_embeddings)
```

```python
use_embeddings = batched_embed(list(data['text']))
```

```python
data['rid'] = data['rid'].combine_first(data['tid'])

most_popular_threads = data['rid'].value_counts().iloc[0:1000].index.values

filtered_data = data[data['rid'].isin(most_popular_threads)]

def to_bow(rids):
    res = np.zeros(1000)
    for rid in rids:
        for i, thread in enumerate(most_popular_threads):
            if rid == thread:
                res[i] += 1
    return res

retweet_embeddings = data.groupby('uid').apply(lambda d: to_bow(d['rid'].values))

retweet_embeddings = retweet_embeddings[~retweet_embeddings.isna()]
```

```python
import tldextract
def extract_domain(url):
    ext = tldextract.extract(url)
    return('.'.join([ext.domain, ext.suffix]))
```

```python
data_user_grouped_urls = data.groupby('uid').apply(lambda d: [extract_domain(e) for l in d['urls'] for e in l])
```

```python
def dummy_fun(doc):
    return doc

tfidf = TfidfVectorizer(
            analyzer='word',
            tokenizer=dummy_fun,
            preprocessor=dummy_fun,
            token_pattern=None,
            min_df = 10)

toy = data_user_grouped_urls[data_user_grouped_urls.apply(len) > 0]

url_embeddings = tfidf.fit_transform(toy.apply(lambda l: [str(s) for s in l])).todense()
```

```python
from sklearn.preprocessing import LabelEncoder
import functools
from sklearn.metrics.pairwise import pairwise_distances_chunked
from sklearn.utils import check_random_state

def silhouette_score_p(
    X, labels, *, metric="euclidean", sample_size=None, random_state=None, **kwds
):
    if sample_size is not None:
#         X, labels = check_X_y(X, labels, accept_sparse=["csc", "csr"])
        random_state = check_random_state(random_state)
        indices = random_state.permutation(X.shape[0])[:sample_size]
        if metric == "precomputed":
            X, labels = X[indices].T[indices].T, labels[indices]
        else:
            X, labels = X[indices], labels[indices]
    return silhouette_samples_p(X, labels, metric=metric, **kwds)

def _silhouette_reduce_p(D_chunk, start, labels, label_freqs):
    # accumulate distances from each sample to each cluster
    clust_dists = np.zeros((len(D_chunk), len(label_freqs)), dtype=D_chunk.dtype)
    for i in range(len(D_chunk)):
        clust_dists[i] += np.bincount(
            labels, weights=D_chunk[i], minlength=len(label_freqs)
        )

    # intra_index selects intra-cluster distances within clust_dists
    intra_index = (np.arange(len(D_chunk)), labels[start : start + len(D_chunk)])
    # intra_clust_dists are averaged over cluster size outside this function
    intra_clust_dists = clust_dists[intra_index]
    # of the remaining distances we normalise and extract the minimum
    clust_dists[intra_index] = np.inf
    clust_dists /= label_freqs
    inter_clust_dists = clust_dists.min(axis=1)
    return intra_clust_dists, inter_clust_dists

def silhouette_samples_p(X, labels, *, metric="euclidean", **kwds):
#     X, labels = check_X_y(X, labels, accept_sparse=["csc", "csr"])

    # Check for non-zero diagonal entries in precomputed distance matrix
    if metric == "precomputed":
        atol = np.finfo(X.dtype).eps * 100
        if np.any(np.abs(np.diagonal(X)) > atol):
            raise ValueError(
                "The precomputed distance matrix contains non-zero "
                "elements on the diagonal. Use np.fill_diagonal(X, 0)."
            )

    le = LabelEncoder()
    labels = le.fit_transform(labels)
    n_samples = len(labels)
    label_freqs = np.bincount(labels)
#     check_number_of_labels(len(le.classes_), n_samples)

    kwds["metric"] = metric
    reduce_func = functools.partial(
        _silhouette_reduce_p, labels=labels, label_freqs=label_freqs
    )
    results = zip(*pairwise_distances_chunked(X, reduce_func=reduce_func, **kwds))
    intra_clust_dists, inter_clust_dists = results
    intra_clust_dists = np.concatenate(intra_clust_dists)
    inter_clust_dists = np.concatenate(inter_clust_dists)
    return(intra_clust_dists,inter_clust_dists)
```

```python
intersection_rt = ground_truth_mask_rt.index.isin(ground_truth_mask_ht.index) &  ground_truth_mask_rt.index.isin(ground_truth_mask_use.index)
intersection_ht = ground_truth_mask_ht.index.isin(ground_truth_mask_rt.index) & ground_truth_mask_ht.index.isin(ground_truth_mask_use.index)
intersection_use = ground_truth_mask_use.index.isin(ground_truth_mask_rt.index) & ground_truth_mask_use.index.isin(ground_truth_mask_ht.index)
```

```python
ground_truth_mask_rt = ground_truth_mask_rt & intersection_rt
ground_truth_mask_ht = ground_truth_mask_ht & intersection_ht
ground_truth_mask_use = ground_truth_mask_use & intersection_use
```

```python
X_rt = np.vstack(retweet_embeddings_df[ground_truth_mask_rt]['retweet_embeddings'].values)
y_rt = retweet_embeddings_df['gt_label'][ground_truth_mask_rt]

X_ht = ht_embeddings[ground_truth_mask_ht]
y_ht = user_ground_truth_df['gt_label'][ground_truth_mask_ht]

X_use = user_use_embeddings[ground_truth_mask_use]
y_use = use_embeddings_df['gt_label'][ground_truth_mask_use]
```

```python
X = np.hstack([X_rt,X_ht,X_use])
y = y_rt

from lightgbm import LGBMClassifier

et = LGBMClassifier(n_estimators=100, min_data_in_leaf=500,colsample_bytree=0.8, class_weight='balanced', n_jobs=-1)

# validation instance
skf = StratifiedKFold(n_splits=2, shuffle=True)

# getting the model validation predictions
preds = cross_val_predict(et, X, y, cv=skf, method='predict_proba')

# evaluating the model
# print('Area under the ROC Curve:', roc_auc_score(y, preds[:,1]))
print('Area under the ROC Curve:', roc_auc_score(y, preds, multi_class='ovo'))
lgbm_performance[feature_set] = roc_auc_score(y, preds, multi_class='ovo')
```
